```python
import pandas as pd

# A침os y meses
a침os = [2017, 2018, 2019, 2020]
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
         "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

# Construir lista de fechas
fechas = []
for a침o in a침os:
    for mes in meses:
        fechas.append(f"{mes} {a침o}")

# Simulaci칩n de valores (reemplazar con reales si los tienes)
datos = []
for fecha in fechas:
    energia = 4.8 + (hash("E" + fecha) % 200) / 100  # simulado
    iqr = 0.6 + (hash("I" + fecha) % 100) / 100
    datos.append({
        "Mes_A침o": fecha,
        "Energia_kWh_m2_dia": round(energia, 2),
        "IQR_kWh_m2_dia": round(iqr, 2),
        "Fuente": "ERA5"
    })

df_energia_era5_boxplot = pd.DataFrame(datos)
df_energia_era5_boxplot.head()
```
